# Chart performance

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

dat = np.genfromtxt('latest_model.csv', delimiter=',', filling_values=np.nan)
rewards = dat[:,0]
baseline = dat[0:33,1]
baseline_mean = baseline.mean()
baseline_std = baseline.std()
print(baseline_mean, baseline_std)
```

```python
plt.plot(range(1,6), rewards[0:5], label='Random')
plt.plot(range(5, len(rewards)+1), rewards[4:], label='SAC')
dummy = np.ones(len(rewards))
plt.plot(range(1,len(rewards)+1), baseline_mean*dummy, c='gray', label='Constant')
plt.fill_between(range(len(rewards)), (baseline_mean-2*baseline_std)*dummy, (baseline_mean+2*baseline_std)*dummy,
                 alpha=0.2, facecolor='gray')
plt.xlabel('Episodes')
plt.ylabel('Delay (vehicle-hours)')
plt.title('Per-episode performance by policy')
plt.legend()
plt.xlim(left=0)
plt.savefig('chart.png', dpi='figure')
plt.show()
```
